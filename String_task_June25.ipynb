#    # Task 1 - Replace string spaces with symbol

```python
def rep_symbol(string):
    new = ''
    new = string.replace(' ','-')
    return new
    print(new)
```

```python
string = 'Hello World this is so much amzing'
rep_symbol(string)
 
```

```python
def rep_symbol(string,sym):
    new = ''
    new = string.replace(' ',sym)
    return new
    print(new)
```

```python
string = 'Hello World this is so much amzing'
sym = '_'
rep_symbol(string,sym)
```

#  Task 2 - List to string

```python
li = ['12','3','4','5','678','90']
def li_to_str(li):
    str = " "
    for i in li:
        str += i
    return str
print(li_to_str(li))
    
```
