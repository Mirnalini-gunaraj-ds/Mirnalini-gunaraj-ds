```python
class one:
    def __init__(self,a1,b1,c1):
        self.a1 = a1
        self.b1 = b1
        self.c1 = c1
        
    def m1(self,a1,b1,c1):
        print("First class has {} ,{} and {}".format(self.a1,self.b1,self.c1))
class two:
    def __init__(self,a2,b2,c2):
        self.a2 = a2
        self.b2 = b2
        self.c2 = c2
        
    def m2(self,a1,b1,c1):
        print("Second class has {} ,{} and {}".format(self.a2,self.b2,self.c2))
    
class three:
    def __init__(self,a3,b3,c3):
        self.a3 = a3
        self.b3 = b3
        self.c3 = c3
        
    def m3(self,a1,b1,c1):
        print("Third class has {} ,{} and {}".format(self.a3,self.b3,self.c3))

class four:
    def __init__(self,a4,b4,c4):
        self.a4 = a4
        self.b4 = b4
        self.c4 = c4
        
    def m4(self,a4,b4,c4):
        print("Fourth class has {} ,{} and {}".format(self.a4.m1(),self.b4.m2(),self.c4.m3()))

a = one(1,2,3)
b = two(4,5,6)
c = three(7,8,9)
d = four(a,b,c)

d.m4()
```

```python
class a:
    def __init__(self,a1,b1):
        self._a1 = a1
        self.__b1 = b1
        print("_ is protected variable")
        print("__ is private variable")
class b(a):
    def show(self):
        print("Class b has {} and {}".format(self._a1,self._a__b1))
obj = b(1,2)
obj.show()


```
