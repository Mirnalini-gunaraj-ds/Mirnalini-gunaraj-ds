# Program to try and catch Exception

```python

def comp(n):
    try:
        a = n/0
    except Exception as E:
        print("Handling with General Exception name , Inside catch block",E)
    finally:
        print("Inside Finally")
comp(5)
```

```python
 #MEthod2
def comp(n):
    try:
        a = n/0
    except ZeroDivisionError:
        print("Handling with System defined Exception name")
    finally:
        print("Inside Finally")
comp(5)
 
```

# Generate Sentence

```python
#Method2
subjects=["Americans ","Indians"]
verbs=["play","watch"]
objects=["Baseball","Cricket"]
li = [i + j + k for i in subjects for j in verbs for k in objects]
print(li)
```

```python
#method1
subjects=["Americans","Indians"]
verbs=["play","watch"]
objects=["Baseball","Cricket"]
for i in subjects:
    for j in verbs:
        for k in objects:
            print(i + ' ' + j + ' ' +k +'.')
```
