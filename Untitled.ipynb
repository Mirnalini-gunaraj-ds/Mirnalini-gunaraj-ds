```python
class a:
    def __init__(self,a):
        self.a = a
    def test(self):
        print("Class A")
class b:
    def __init__(self,b):
        self.a = b
    def test(self):
        print("Class B")
class c(a,b):
    obj1 = a(1)
    obj2 = b(2)
    obj1.test()
    obj2.test()  
            
```

```python
class a:
    def test1(self):
        print("Class A")
class b(a):
    def test2(self):
        print("Class B")
class c(b):
    def test3(self):
        print("Class C")
        
obj = c()
obj.test1()
obj.test2()
```
