# 5) Count all lower case, upper case, digits, and special symbols

```python
s1 = '3333'

low = up = dig = pr = 0

for i in s1:
    if i.islower():
        low = low + 1
    elif i.isupper():
        up = up + 1
    elif i.isdigit():
        dig = dig + 1
    elif i.isprintable():
        pr = pr + 1
print(low)
print(up)
print(dig)
print(pr)
        
```
