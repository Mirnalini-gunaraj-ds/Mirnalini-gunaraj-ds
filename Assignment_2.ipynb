```python
# 2) Assignment_2
# 1) Pattern creation

def pattern(i):
    for j in range(i):
        print('*',end = " ")
    print("\n")
for i in range(1,6):
    pattern(i)
for i in range(4,0,-1):
    pattern(i)    
```
