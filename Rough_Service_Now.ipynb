```python
#creating dataframe from dict
import pandas as pd
import re
pd.options.display.max_colwidth=100

vals = []
for j in data_co['result']:
    cols = j.keys()
    vals.append(j.values())
sn = pd.DataFrame(data=vals,columns = cols)
sn
```

```python
#Extracting screen name from description

def func_ex(pat,lst):
    sub_lst = []
    sear = re.findall(pat,lst,re.IGNORECASE)
    if sear == None:
        return np.nan
    else:
        sub_lst = [i for i,j in sear]
        return sub_lst
pattern1 = r'((sg)[a-zA-Z]{2})'
sn['Scr_new'] = sn['sd'].apply(lambda i:func_ex(pattern1,i))
```

```python
#Extracting error codefrom description

def func_ex1(pat,lst):
    sub_lst1 = []
    sear = re.findall(pat,lst)
    if sear == None:
        return np.nan
    else:
        return sear
pattern2 = r'[0-9]{4}[R|r]'
sn['err_new'] = sn['sd'].apply(lambda i:func_ex1(pattern2,i))
```

```python
#Capitalise elements in list
def capill(r):
    newz=[]
    map_object = map(str.upper,r)     
    newz = list(map_object)
    return newz

sn.Scr_new = sn.Scr_new.apply(capill)
sn.err_new = sn.err_new.apply(capill)

sn1 = sn.copy()
sn1
```

```python
#sn1['corp'] = sn1['Scr_new'].apply(lambda l:l) + sn['err_new'].apply(lambda l:l)
#corps = sn1['corp'].map(' '.join)

# Function to combine screen code and error code
def combo(combi):
    combi['corp'] = combi['Scr_new'].apply(lambda l:l) + combi['err_new'].apply(lambda l:l)
    
combo(sn1)
corps = sn1['corp'].map(' '.join)
```

```python
#Creating BOW
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corps)
print(cv.vocabulary_)
```

```python
#print(X)
```

```python
#creating tf-idf
from sklearn.feature_extraction.text import TfidfTransformer
tfi = TfidfTransformer().fit(X)
#print(tfi)

tfid = tfi.transform(X)

```

```python
#importing libraries and creating objects
import nltk
from nltk.corpus import stopwords

from nltk.stem import PorterStemmer
ps=PorterStemmer()

from nltk.stem import WordNetLemmatizer 
ls= WordNetLemmatizer()

import re
```

```python
#Cleaning y variable
corpy = []
for e in range(0,len(sn1)):
    resol = re.sub('[^a-zA-Z]',' ',sn1['resolution'][e])
    resol = resol.lower()
    resol = resol.split()
    #resol = [ps.stem(word) for word in resol if not word in stopwords.words('english')]
    resol = [ls.lemmatize(word) for word in resol if not word in stopwords.words('english')]
    #print(resol)
    resol = ' '.join(resol)
    #print(resol)
    corpy.append(resol)
    #print(corpy)
```

```python
sn1['labels'] = corpy
sn1
```

```python
#Taking label variable to y
y = sn1['labels']
```

```python
#Train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(tfid,y,test_size=0.50,random_state=0)
#print("X_train \n",X_train)
#print("y_train \n",y_train)
#print("X_test \n",X_test)
#print("y_test \n",y_test)
```

```python
#Applying Naive bayes Model

from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB().fit(X_train,y_train)
```

```python
#Predicting using Naive Bayes model
predic = model.predict(X_test)
```

```python
#Accuracy

from sklearn.metrics import confusion_matrix,classification_report
matrix1 = confusion_matrix(y_test,predic)
matrix2 = classification_report(y_test,predic)
print(matrix1)
print(matrix2)
```
