```python
# Fitting periodic function

import numpy as np
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17,18
Date = np.arange(1,13)
tMax = np.array(Max)
tMin = np.array(Min)
```

```python
import matplotlib.pyplot as plt

plt.plot(Date,tMax,'ro')
plt.plot(Date,tMin,'bo')

plt.xlabel('Month')
plt.ylabel('Temperatures(^C)')
plt.show()
```

```python
from scipy import optimize

def temp_yr(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(temp_yr, Date,tMax, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(temp_yr, Date,tMin, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(Date, tMax, 'ro')
plt.plot(days, temp_yr(days, *res_max), 'r-')
plt.plot(Date, tMin, 'bo')
plt.plot(days, temp_yr(days, *res_min), 'b-')
plt.xlabel('Month')
aa= u'\N{DEGREE SIGN}'
plt.ylabel('Temperature {} C'.format(aa))

plt.show() 
```

```python
# Pie chart
import pandas as pd
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic.head()
```

```python
p = titanic['sex'].value_counts()
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(6,6))

plt.title("Male/Female Pie Chart")
color = ['Red','Blue']
plt.pie(p,labels=p.index,autopct='%1.1f%%',colors=color)
plt.show()
```

```python
colors = {'male' : 'red' , 'female' : 'blue'}
plt.scatter(titanic['age'],titanic['fare'], c=titanic['sex'].iloc[:-1].apply(lambda x : colors[x]))
plt.title('Scatter Plot of Fare paid and Age')
plt.xlabel('Age')
plt.ylabel('Fare Paid')
plt.show()
```
