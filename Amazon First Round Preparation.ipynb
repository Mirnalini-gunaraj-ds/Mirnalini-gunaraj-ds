```python
# First find the type-ordinal or nominal ; 
# Ordinal - Label Encoding (Sequence number is added for each category) 
# Nominal - OneHot(Uses 3 digits and creates n columns with binary) , 
# Nominal - Dummy(same as onehot but USes 2 digits and one variable is reduced) ,
# Effect - Same as dummy but uses -1,0,1
# HAsh - Hash Alg like MD5 will be app;ied and the no of dimensions can be selected but data is lost and also collision
#             Multiple values have same hash value is called collision
# Binary - Combination of one hot and Hash (cat -to ordinal to binary) suitable for large categories
# BaseN - Same like binary but we can pass the base value as 10 for decimal and respective number for octal
# Target - uses the mean value of the targle variable for each category and assigns to each category instead of 0 and 1 but 
#              leads to overfitting of the model
```
