```python
#Mapping Yes as 1 and No as 0
rein_dict = {"Yes": 1 , "No" : 0}
def func(x,rein_dict):
    try:
        tmp = rein_dict[x]
        return(tmp)
    except:
        return('default')
    
df['Reinstated'] = df['Reinstated'].map(lambda x : func(x,rein_dict))
```

# Mean Aboslute error function

```python
#Retrieving object datatype alone

s= (X_train.dtypes=='object')
obj_cols = list(s[s].index) #s[s] returns only True variables alone
len(obj_cols)


#Measure quality of each approach

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_ds(X_trains,X_valids,y_trains,y_valids):
    model = RandomForestRegressor(n_estimators=200,random_state=0)
    model.fit(X_trains,y_trains)
    pred = model.predict(X_valids)
    return mean_absolute_error(y_valids,pred)
```
