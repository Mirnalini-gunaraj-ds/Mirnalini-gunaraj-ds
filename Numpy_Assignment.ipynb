```python
#Write a function to find moving average in an array over a window:
inp = np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])
k = 3
n = np.size(inp)
mov = (n-k+1)
mov
for i in range(0, mov):
    fo = (inp[i:k+1]) 
    an = sum(fo)/4
    k = k+1
    print("Y{}".format(i+1) , ":" , an)
```
