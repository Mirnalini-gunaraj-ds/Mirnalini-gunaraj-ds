# 1) Print numbers

```python
li = []
for i in range(2000,3201):
    if (i % 7 == 0) and (i % 5 != 0):
        li.append(i)
print("The numbers in sequence order which are divisible by 7 and not multiple of 5 are \n" , li)
print("The total number of elements which satisfy the condition is " , len(li))
```

# 2.a) Reversing name fully

```python
print("Please enter your first name and last name")
name = input()
print("The normal order is " + name)
print("The reversed order taking full name into account " + name[::-1])
```

# 2.b)Reversing separately

```python
print("Enter your first name")
first = input()
print("Enter your last name")
last = input()
print("The reversed order is {}".format(first[::-1] + ' ' + last[::-1]))
```

# 3) Calculate Volume

```python
import math
phi = math.pi

diameter = 12
radius = int(diameter / 2)

def calc(radius):
    vol = (4/3 * phi * radius )
    print("The volume is " , vol)
calc(radius)
```
