# 1.1 ) Area of triangle - Class 

```python
class A:
    def __init__(self,s,a,b,c):
        self.s = s
        self.a = a
        self.b = b
        self.c = c
        print("The values of triangle are ",self.s,self.a,self.b,self.c)
    
class B(A):
    def calc(self):
        area = (self.s * (self.s - self.a) *  (self.s -self.b)  * (self.s - self.c)) ** 0.5 
        print("The area is", area)
obj2 = B(36,18,24,30)
obj2.calc()
```

# 1.2) Print Words longer than given length

```python

# Using Normal Function
def filter_long_words(li,n):
    for i in li:
        if (len(i) > n):
            lo.append(i)
    return lo
li = ['Gratefulness','Blissful','wow','nature','world','is','heaven','Lovely']
n = 5
lo = []
print(filter_long_words(li,n))

#using filter function:
li = ['Gratefulness','Blissful','wow','nature','world','is','heaven','Lovely']
n = 9
fil =  list(filter(lambda i : (len(i)>n), li))
print(fil)
```

# 2.2 ) Vowel Check

```python
def vow(c):
    vowel = ['a','e','i','o','u']
    if c in vowel:
        return True
    else:
        return False
    
c = input('Please enter a character')
if (len(c) == 1):
    print(vow(c))
else :
    print('Invalid character. Please enter a single character')
```
